```python
''' So the approach I used is to first create a dataframe which is the monthly milk prod.
  After that I divided the dataframe into 12 data frames each representing a month and stored
  the data of each month in a dictionary. Data of each month can be accessed using the month name
  each month name is a key of the dictionary, nd the value corresponding to it is the data of that month.

  '''
import pandas as pd

df = pd.read_excel('/content/Dairy Economics_Master Sheet.xlsx', sheet_name='Monthly Milk Prod, Cons. & Sale')
df.head()
```

```python
for x in df.columns:
  df.drop(columns=[x],inplace=True)
  break
# I used this to drop the S.No column, for some reason I was not gettin the exact name of the column, so had to use this approach
```

```python
months=["Jan","Feb","March","April","May","June","July","Aug","Sep","Oct","Nov","Dec"]
counter=0
months_data={}
for x in months:
  months_data[x] = df.iloc[:, counter:counter + 16].apply(pd.to_numeric, errors='coerce').fillna(0)
  counter+=16 # HEre we are using 16 because each month's data spans over 16 columns
months_data["Feb"]
```

```python
# Let's get the monthly data of cow and buffalo -> Total milk production
cow={}
buffalo={}
for x in months:
  cow[x]=months_data[x].iloc[4:,0:1]
  buffalo[x]=months_data[x].iloc[4:,1:2]
cow["Jan"]
```

```python
buffalo["Jan"]
```

```python
cow["Feb"]
```

```python
# Now we will calculate monthly standard deviation.
import numpy as np
std_cow = {}
std_buffalo = {}

for x in months:
    cow_filtered = cow[x].replace(0, np.nan).dropna()
    buffalo_filtered = buffalo[x].replace(0, np.nan).dropna()
    std_cow[x] = np.std(cow_filtered)
    std_buffalo[x] = np.std(buffalo_filtered)
```

```python
import matplotlib.pyplot as plt

# Extract standard deviation values for plotting
months = ["Jan", "Feb", "March", "April", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]
std_cow_values = [std_cow[month].values[0] for month in months]
std_buffalo_values = [std_buffalo[month].values[0] for month in months]

# Plotting the standard deviation values
plt.figure(figsize=(12, 6))
plt.plot(months, std_cow_values, marker='o', label='Cow', linestyle='-', linewidth=2)
plt.plot(months, std_buffalo_values, marker='s', label='Buffalo', linestyle='--', linewidth=2)

# Adding titles and labels
plt.title('Monthly Standard Deviation of Milk Production')
plt.xlabel('Month')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
